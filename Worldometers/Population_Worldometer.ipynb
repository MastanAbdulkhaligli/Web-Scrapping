```python
from selenium import webdriver
import pandas as pd
```

```python
driver = webdriver.Chrome(r'C:\Users\User\Desktop\chromedriver.exe')

driver.get('https://www.worldometers.info/world-population/population-by-country/')
driver.maximize_window()

```

```python
countries = driver.find_elements_by_xpath('//tbody/tr/td/a')
population = driver.find_elements_by_xpath('//tbody/tr/td[3]')
yearly_change = driver.find_elements_by_xpath('//tbody/tr/td[4]')
net_change = driver.find_elements_by_xpath('//tbody/tr/td[5]')
density = driver.find_elements_by_xpath('//tbody/tr/td[6]')
land_area = driver.find_elements_by_xpath('//tbody/tr/td[7]')
migrants=driver.find_elements_by_xpath('//tbody/tr/td[8]')
fert_rate=driver.find_elements_by_xpath('//tbody/tr/td[9]')
med_age=driver.find_elements_by_xpath('//tbody/tr/td[10]')
urban_pop=driver.find_elements_by_xpath('//tbody/tr/td[11]')
world_share =driver.find_elements_by_xpath('//tbody/tr/td[12]')

table = []

for i in range(len(countries)):
    data_dict = { "Country" : countries[i].text,
                 "Population": population[i].text,
                 "Yearly Change": yearly_change[i].text,
                 "Net Change": net_change[i].text,
                 "Density(P/KM)": density[i].text,
                 "Land Area": land_area[i].text,
                 "Migrants": migrants[i].text,
                 "Fert Rate": fert_rate[i].text,
                 "Med. Age": med_age[i].text,
                 "Urban Pop": urban_pop[i].text,
                 "World Share)": world_share[i].text
    }
    
    table.append(data_dict)
    
len(table)


```

```python
df = pd.DataFrame(table) 
df
```

```python
df.to_csv("Worldometers.csv", encoding = 'utf-8',index = False)
```
