## Web Scrapping from www.basketball-reference.com

### Automatically return a list of URL given a list of years 
Original link https://www.basketball-reference.com/leagues/NBA_2021_per_game.html
* I can use .format with {} and give input as a list of years.
* That is why i define link variable which is string and put {} inside 
* My years list will include all years from 2000 to 2021
* To define this list i will use "List Comprehension" you can use np.arrange and np.fill method also
* If u want to learn about list comprehension you can check List Comprehension in my github
* After using .format i append all my str(link address) to my link_list to make easy when loading data

```python
years = [ i for i in range (2000,2022)]
link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
link_list = []

for year in years:
    url= link.format(year)
    print(url)
    link_list.append(url)
```

## Now lets import some libraries of course Pandas comes first :)


```python
import pandas as pd

url1 = 'https://www.basketball-reference.com/leagues/NBA_2000_per_game.html'
```

### First I define df1 variable and read 2000 data outside of for loop to show you how to do this then i got all data from my years list and append them into one dataframe
* Important note when we call read_html() function it returns list not DataFrame that is why after that we should convert it to DataFrame using [0] value of list
* In addition to that, I add year column to my data because we combine all of data 

```python
df1 = pd.read_html(url1,header =0)
df1 = pd.DataFrame(df1[0])
df1["Year"] = years[0]


for item in range(len(link_list)-1):
    temp = pd.read_html(link_list[item+1],header = 0)
    temp = pd.DataFrame(temp[0])
    temp["Year"] = years[item+1]
    df1 = df1.append(temp, ignore_index= True)
    
df1.to_csv("All_NBA_Data.csv",encoding='utf-8')
```
